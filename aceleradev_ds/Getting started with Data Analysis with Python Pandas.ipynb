## Preenchendo valores faltantes com .fillna()
- Para preencher valores faltante no dataframe, existe um métado chamado .fillna().
- Vamos assumir que nos temos muitos valores faltando na coluna e nos queremos preencher com zero.
- df["columnname"].fillna(0,inplace=True)
- Mas em vez de preencher com 'Unknown' preenchemos o valores faltante da coluna Age, vamos usar a mediana da coluna

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

### Checando se aida existe algum valor null na coluna Age

```python
df['Age'].isnull().sum()
```
